#Linear Regression predicts continuous values, which aligns well with the nature of many regression tasks. If the target variable (like OSI) has a continuous and linear relationship with the features, Linear Regression can accurately capture and predict those trends, resulting in high R values.

In contrast, Decision Trees and Random Forests work by segmenting the data into discrete splits. This makes them better suited for capturing non-linear patterns or interactions between features. However, if the underlying relationship is linear, these models may struggle, leading to poor predictions and low R values. Their discrete approach doesn't capture smooth trends as well as Linear Regression.
